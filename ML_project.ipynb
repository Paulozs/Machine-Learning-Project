### read_dataset(datapath, labeled)
- datapath: "./Data/whic folder/which file"
- labeled: If True returns {"label":[all emissions]}, If False returns [all emissions]

```python
##if labelled = true returns a dict
def read_dataset(datapath, labeled=True):
    with open(datapath, encoding="utf-8") as data:
        data_read = data.read()
    states = {}
    state_emission = data_read.split("\n")
    filtered_state_emission= [word for word in state_emission if word.strip()!= ""]
    results = filtered_state_emission
    if(labeled):
        pairs = [tuple(pair.split()) for pair in filtered_state_emission]
        for pair in pairs:
            if pair[1] not in states.keys():
                states.update({pair[1]:[pair[0]]})
            elif pair[1] in states.keys():
                states[pair[1]].append(pair[0])
        results = states
    return results

ES_train = read_dataset("./Data/ES/train")
ES_dev = read_dataset("./Data/ES/dev.in",False)
ES_devout = read_dataset("./Data/ES/dev.out")
print(ES_train)
print(ES_dev)
print(ES_devout)
```
